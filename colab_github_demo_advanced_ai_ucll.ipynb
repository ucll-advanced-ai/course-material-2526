## Using Google Colab for neural network training

This guide provides step-by-step instructions on how to clone a GitHub repository to Google Drive and work with it using Google Colab. Google Colab is a powerful tool that allows you to run Python code in the cloud and interact with your Google Drive files. There are two parts:
  1) the one time setup and
  2) some parts you always have to add at the top of every notebook

## Prerequisites

1. A Google account with access to Google Drive.
2. A GitHub account.
3. A GitHub personal access token.

## One time setup: clone the repository to google drive (and do a git pull when there are new changes )

### 1. Mount Google Drive

First, mount your Google Drive to the Colab environment. This allows you to access and save files directly to your Google Drive.

When you run this code, you will have to authorize Google Drive access. Follow the instructions to complete the authorization process.

```
from google.colab import drive
drive.mount('/content/drive')
```

### 2. Configure Git

Set up your Git configuration with your user information.

```
!git config --global user.name "user name "
!git config --global user.email "username@mail.com"
```

### 3. Generate a GitHub Personal Access Token

To authenticate with GitHub, generate a personal access token:

1. Go to your [GitHub settings](https://github.com/settings/tokens).
2. Click "Generate new token".
3. Select the appropriate scopes (usually `repo`).
4. Copy the generated token.

### 4. Clone the Repository

Use the following code to clone your repository to Google Drive. Replace `your_token`, `your_username`, and `your_repo_name` with your actual details.

```
token = 'token'
username = 'username'

!git clone https://{username}:{token}@github.com/ucll-advanced-ai/course-material-2526.git /content/drive/MyDrive/GitHub/course-material-2526
```


### 5. Navigate to the Repository

Change the directory to your cloned repository. You should now be able to execute typical github commands `git status` or `git pull` etc.

## how to use the jupyter notebooks in google colab

### 1) Open the notebook
- navigate to the notebook you want to open (in Google Drive) and open the notebook. It should automatically open in google Colab.
- You can directly start with 02_neural_networks.ipynb.
- The setup described in 01_setup is only necessary if you work local. Google Colab provides Python versions where all packages are already installed

### 2) Choose your runtime
on the topright, click on the arrow next to 'RAM' and 'disk', and click on `change runtime type`

Choose one of the available GPUs. in the free version, GPUs will be available for a limited time, and the RAM and disk space is also limited, but it will probably still run a lot better than on your computer if you don't have access to a GPU yourself.
